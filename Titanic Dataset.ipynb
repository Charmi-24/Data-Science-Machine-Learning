```python
titanic1.drop("Cabin",axis=1,inplace= True)
titanic1.head(5)
```

```python
titanic1.drop("Age",axis=1,inplace= True)
titanic1.head(3)
```

```python
sns.heatmap(titanic1.isnull(),yticklabels = False,cbar = False)
```

```python
titanic1.dropna(inplace= True)
# na is used because we want to drop null values only and not a whole column
```

```python
titanic1.isnull().sum()
```

```python
# pandas' function
var=pd.get_dummies(titanic1['Sex'])
var.head(2)
```

```python
var=pd.get_dummies(titanic1['Sex'],drop_first=True)
var.head(2)
```

```python
var1=pd.get_dummies(titanic1['Embarked'],drop_first=True)
var1.head(2)
```

```python
var2=pd.get_dummies(titanic1['Pclass'],drop_first=True)
var2.head(2)
```

```python
titanic1=pd.concat([titanic1,var,var1,var2],axis=1)
titanic1.head(5)
```

```python
titanic1.drop(["PassengerId","Name","Sex","Pclass","Ticket","Embarked"],axis=1,inplace= True)
titanic1.head(5)
```

```python
X=titanic1.drop("Survived",axis=1)
Y=titanic1["Survived"]
print(X)
print(Y)
```

# Algorithms

# K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
object1=KNeighborsClassifier(n_neighbors=5)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25)
object1.fit(X_train,Y_train)
Y_predict1=object1.predict(X_test)
print(Y_predict1[0])
print(Y_test)
```

```python
acc1_knn=metrics.accuracy_score(Y_test,Y_predict1)*100

print("K nearest neighbors accuracy (in %):",acc1_knn)
```

# Logistic Regression

```python
from sklearn import linear_model
object2=linear_model.LogisticRegression()
object2.fit(X_train,Y_train)
Y_predict2=object2.predict(X_test)
print(Y_predict2[0])
print(Y_test)
```

```python
acc2_lr=metrics.accuracy_score(Y_test,Y_predict2)*100
print("Logistic Regression model accuracy (in %):",acc2_lr)
```

# Support Vector Machine

```python
from sklearn import svm
object3=svm.SVC(kernel='linear')
object3.fit(X_train,Y_train)
Y_predict3=object3.predict(X_test)
print(Y_predict3[0])
print(Y_test)
```

```python
acc3_svm=metrics.accuracy_score(Y_test,Y_predict3)*100
print("SVM accuracy (in %):",acc3_svm)
```

```python
from matplotlib import style
style.use('ggplot')
plt.plot(Y_predict3)
plt.show()
```

```python
names=['KNN','LR','SVM']
values=['Y_predict1','Y_predict2','Y_predict3']
plt.figure(1,figsize=(20,5))
plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.plot(names,values)
plt.show()
```

# GaussianNB (naives_bayes)

```python
from sklearn.naive_bayes import GaussianNB
object4 = GaussianNB()
Y_predict4 = object4.fit(X_train,Y_train).predict(X_test)
print(Y_predict4[0])
print(Y_test)
```

```python
acc4_GNB=metrics.accuracy_score(Y_test,Y_predict4)*100
print("GNB accuracy (in %):",acc4_GNB)
```

# MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
object5 = MultinomialNB()
Y_predict5 = object5.fit(X_train,Y_train).predict(X_test)
print(Y_predict5[0])
print(Y_test)
```

```python
acc5_MNB=metrics.accuracy_score(Y_test,Y_predict5)*100
print("MNB accuracy (in %):",acc5_MNB)
```

# BernoulliNB

```python
from sklearn.naive_bayes import BernoulliNB
object6 = BernoulliNB()
Y_predict6 = object6.fit(X_train,Y_train).predict(X_test)
print(Y_predict6[0])
print(Y_test)
```

```python
acc6_BNB=metrics.accuracy_score(Y_test,Y_predict6)*100
print("BNB accuracy (in %):",acc6_BNB)
```

# Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier 
object7 = DecisionTreeClassifier(criterion="gini")
Y_predict7 = object7.fit(X_train,Y_train).predict(X_test)
print(Y_predict7[0])
print(Y_test)
```

```python
acc7_DT=metrics.accuracy_score(Y_test,Y_predict7)*100
print("DT accuracy (in %):",acc7_DT)
```
